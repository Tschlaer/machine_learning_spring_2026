```python
# For columns where the NA values show the features are not included, changes the NA value to None
none_cols = [
    "Alley","PoolQC","Fence","MiscFeature","FireplaceQu",
    "GarageType","GarageFinish","GarageQual","GarageCond",
    "BsmtQual","BsmtCond","BsmtExposure","BsmtFinType1","BsmtFinType2",
    "MasVnrType"
]
for i in none_cols:
    if i in df.columns:
        df[i] = df[i].fillna("None")
```

```python
# Change LotFrontage missing values according to neighborhood median
df["LotFrontage"] = df.groupby("Neighborhood")["LotFrontage"]\
                      .transform(lambda x: x.fillna(x.median()))
df["LotFrontage"] = df["LotFrontage"].fillna(df["LotFrontage"].median())
```

```python
# Fill missing value for Electrical with mode
df["Electrical"] = df["Electrical"].fillna(df["Electrical"].mode()[0])
```

```python
# Check Missing Values
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
# Addressing significant ouliers using winsorization
def cap_outliers(df, col):
    # IQR bounds
    x = df[col]
    q1, q3 = x.quantile([0.25, 0.75])
    iqr = q3 - q1
    low, high = q1 - 1.5*iqr, q3 + 1.5*iqr

    # Cap outliers
    df[col] = np.where(df[col] < low,  low,
                np.where(df[col] > high, high, df[col]))

    return low, high

# Apply to all five variables
columns_to_cap = ["LotArea", "GrLivArea", "TotalBsmtSF", "1stFlrSF", "GarageArea"]
bounds = {}

for col in columns_to_cap:
    low, high = cap_outliers(df, col)
    bounds[col] = (low, high)

bounds
```

```python
# Higlights high correlation pairs which exhibit multicollinearity
num_df = df.select_dtypes(exclude="object")
corr = num_df.corr()

upper = corr.where(np.triu(np.ones(corr.shape), k=1).astype(bool))

high_corr_pairs = (
    upper.stack()
         .reset_index()
         .rename(columns={"level_0":"feature_1", "level_1":"feature_2", 0:"corr"})
)

high_corr_pairs = high_corr_pairs.assign(abs_corr = high_corr_pairs["corr"].abs())
high_corr_pairs = high_corr_pairs.query("abs_corr >= 0.80").sort_values("abs_corr", ascending=False)

high_corr_pairs
```
